```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds=pd.read_csv("diamonds.csv")
diamonds.head()
```

```python
sns.distplot(diamonds['price'])
plt.show()
```

```python
# if u only want to see frequency polygon and not histogram
sns.distplot(diamonds['price'],hist=False)
plt.show()
```

```python
sns.distplot(diamonds['price'],color='red')
plt.show()
```

```python
# if u only want to see histogram and not frequency polygon
sns.distplot(diamonds['price'],kde=False)
plt.show()
```

```python
#to plot vertically
sns.distplot(diamonds['price'],vertical=True)
plt.show()
```
