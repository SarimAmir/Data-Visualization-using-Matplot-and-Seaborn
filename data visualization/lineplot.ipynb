```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
plt.plot(x,y,color='b',linestyle="-",linewidth=5)
plt.title("Line plot")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
y1=2*x
y2=3*x
plt.plot(x,y1,color="r",linestyle=":",linewidth=5)
plt.plot(x,y2,color="y",linestyle="-",linewidth=3)
plt.grid(True)
plt.title("Two lines in line plot")
plt.xlabel("x axis")
plt.ylabel("y label")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color="r",linestyle=":",linewidth=5)
plt.title("Two line plot in 2 rows")
plt.xlabel("x axis")
plt.ylabel("y label")

plt.subplot(1,2,2)
plt.plot(x,y2,color="y",linestyle="-",linewidth=3)

plt.grid(True)
plt.title("Two line plot in 2 rows")
plt.xlabel("x axis")
plt.ylabel("y label")
plt.show()
```
