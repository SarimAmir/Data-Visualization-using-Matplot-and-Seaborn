```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
iris=pd.read_csv("iris.csv")
iris.head()
```

```python
sns.scatterplot(x='sepallength',y='petallength',data=iris)
plt.show()
```

```python
#using hue attribute
sns.scatterplot(x='sepallength',y='petallength',data=iris,hue='class')
plt.show()
```

```python
#using style attirubute
sns.scatterplot(x='sepallength',y='petallength',data=iris,hue='class',style='class')
plt.show()
```

```python
#using hue attribute to numerically categorize the values
sns.scatterplot(x='sepallength',y='petallength',data=iris,hue='petallength')
plt.show()
```
