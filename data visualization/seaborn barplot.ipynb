```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
pokemon=pd.read_csv("pokemon.csv")
pokemon.head()
```

```python
sns.barplot(x='is_legendary',y='speed',data=pokemon)
plt.show()
```

```python
sns.barplot(x='is_legendary',y='weight_kg',data=pokemon)
plt.show()
```

```python
sns.barplot(x='is_legendary',y='speed',data=pokemon, hue='generation')
plt.show()
```

```python
#using diffrent palattes
sns.barplot(x='is_legendary',y='speed',data=pokemon,palette='rocket')
plt.show()
```

```python
#using diffrent palattes
sns.barplot(x='is_legendary',y='speed',data=pokemon,palette='vlag')
plt.show()
```

```python
#using diffrent palattes
sns.barplot(x='is_legendary',y='speed',data=pokemon,palette='Blues_d')
plt.show()
```

```python
#using your color of choice
sns.barplot(x='is_legendary',y='speed',data=pokemon,color="red")
plt.show()
```
