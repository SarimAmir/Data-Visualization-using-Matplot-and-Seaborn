```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=[4,2,5,7,2,5,6,8]
y=[9,3,5,1,8,3,4,7]
```

```python
plt.scatter(x,y)
plt.title("Scatter plot")
plt.xlabel("X label")
plt.ylabel("y label")
plt.show()
```

```python
#changing color of marker(dots points)
plt.scatter(x,y,marker='*',color='y',s=30)
plt.title("Scatter plot")
plt.xlabel("X label")
plt.ylabel("y label")
plt.show()
```

```python
#creating 2 markers in single scatter plot
z=[3,5,1,6,8,9,10,4]
plt.scatter(x,y,marker='*',color='y',s=100)
plt.scatter(x,z,marker='.',color="red",s=100)
plt.title("Two markers in a single Scatter plot")
plt.xlabel("X label")
plt.ylabel("y label")
plt.show()
```

```python
#creating subplots
plt.subplot(1,2,1)
plt.scatter(x,y,marker='*',color='y',s=100)
plt.title("plot 1")
plt.xlabel("X label")
plt.ylabel("y label")

plt.subplot(1,2,2)
plt.scatter(x,z,marker='.',color="red",s=100)
plt.title("plot 2")
plt.xlabel("X label")
plt.ylabel("y label")
plt.show()
```
