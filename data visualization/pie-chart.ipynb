```python
from matplotlib import pyplot as plt
```

```python
fruit=['Apple','Orange','Mango','Guava']
quantity=[25,67,100,70]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
#to add percentages in pie chart
plt.pie(quantity,labels=fruit,autopct="%0.1f%%",colors=['yellow','red','blue','green'])
plt.show()
```
