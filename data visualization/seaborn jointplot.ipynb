```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
iris=pd.read_csv("iris.csv")
iris.head()
```

```python
sns.jointplot(x='sepallength',y='petallength',data=iris)
plt.show()
```

```python
sns.jointplot(x='sepallength',y='petallength',data=iris,color='brown')
plt.show()
```

```python
#to add regression line in jointplot wrt scatter and histogram
sns.jointplot(x='sepallength',y='petallength',data=iris,kind="reg")
plt.show()
```
