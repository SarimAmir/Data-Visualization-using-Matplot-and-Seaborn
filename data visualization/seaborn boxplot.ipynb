```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
churn=pd.read_csv("churn.csv")
churn.head()
```

```python
sns.boxplot(x='Churn',y='tenure',data=churn)
plt.show()
```

```python
sns.boxplot(x='InternetService',y='MonthlyCharges',data=churn)
plt.show()
```

```python
#using palette attriubute
sns.boxplot(x='InternetService',y='tenure',data=churn,palette='Set1')
plt.show()
```

```python
#using color attribute

sns.boxplot(x='InternetService',y='tenure',data=churn,color='red')
plt.show()
```

```python
# to change the width of outline of boxplot
sns.boxplot(x='Contract',y='tenure',data=churn,color='red',linewidth=6)
plt.show()
```

```python
# to change the order of categories
sns.boxplot(x='Contract',y='tenure',data=churn,order=["One year Contract","Month-to-month","Two year"])
plt.show()
```

```python
#using hue attribute
sns.boxplot(x='Contract',y='tenure',data=churn,hue='PaymentMethod')
plt.show()
```
